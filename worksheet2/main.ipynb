```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('road_accidents.csv')
```

```python
df
```

```python
df['total-accidents'] = df[['Fine/Clear - Total Accidents', 'Mist/ Foggy - Total Accidents', 'Cloudy - Total Accidents', 'Rainy - Total Accidents',
    'Snowfall - Total Accidents', 'Hail/Sleet - Total Accidents', 'Dust Storm - Total Accidents', 'Others - Total Accidents']].sum(axis=1)
```

```python
df[['State/ UT','total-accidents']]
```

```python
df['total-deth'] = df[['Fine/Clear - Persons Killed', 'Mist/ Foggy - Persons Killed', 'Cloudy - Persons Killed',
     'Rainy - Persons Killed', 'Snowfall - Persons Killed', 'Hail/Sleet - Persons Killed',
     'Dust Storm - Persons Killed', 'Others - Persons Killed']].sum(axis=1)
```

```python
df[['State/ UT','total-accidents','total-deth']]
```

```python
df['fatality-rate'] = df['total-deth']/df['total-accidents']
```

```python
df[['State/ UT','total-accidents','total-deth','fatality-rate']]
```

```python
df.head()
```

```python
df['mist/ Fooggy rate'] = df['Mist/ Foggy - Total Accidents']/df['Mist/ Foggy - Total Accidents'].sum()
```

```python
df['mist/ Fooggy rate']
```

```python
correlation_coefficient, p_value = stats.pearsonr(df['fatality-rate'], df['mist/ Fooggy rate'])
print(correlation_coefficient, p_value )
```

```python
if p_value<0.05 :
    print("there is a significant correlation")
else :
    print("no significant relation")
```

```python
df['Hail/sleet-bin'] = df['Hail/Sleet - Total Accidents'].apply(lambda x:1 if(x>0) else 0)
```

```python
correlation_coefficient, p_value = stats.pearsonr(df['Hail/sleet-bin'], df['Rainy - Total Accidents'])
print(correlation_coefficient, p_value )
```

```python
if p_value<0.05 :
    print("there is a significant correlation")
else :
    print("no significant relation")
```

```python

correlation_coefficient, p_value = stats.pointbiserialr(df['Hail/sleet-bin'], df['Rainy - Total Accidents'])
print(correlation_coefficient, p_value )
```

```python
if p_value<0.05 :
    print("there is a significant correlation")
else :
    print("no significant relation")
```

```python
df['dust/storm bin'] = df['Dust Storm - Total Accidents'].apply(lambda x:1 if(x>0) else 0)
```

```python
df['dust/storm bin']
```

```python
table = pd.crosstab(df['dust/storm bin'],df['Hail/sleet-bin'])
```

```python
a = table.iloc[0, 0] 
b = table.iloc[0, 1] 
c = table.iloc[1, 0] 
d = table.iloc[1, 1] 
numerator = (a*d - b*c)
denominator = np.sqrt((a+b)*(c+d)*(a+c)*(b+d))
phi = numerator / denominator
print(phi)
```

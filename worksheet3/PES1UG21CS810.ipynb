```python
f_stats , p_val = stats.f_oneway(df['A'],df['B'],df['C'],df['D'])
alpha = 0.05
if p_val < alpha:
    print("There is evidence to suggest a difference in average marks among fitness plans.")
else:
    print("There is no significant evidence to suggest a difference in average marks among fitness plans.")
```
